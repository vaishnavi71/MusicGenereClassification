## Feature Engineering

Need to convert the categorical variables into numbers (in short one hot encoding), using LabelEncoder() which encodes each unique label with a digit sequentially

##### The categorical variable / Target Variable/Classes

```python
class_list=df.iloc[:,-1]
converter=LabelEncoder()
```

```python
y=converter.fit_transform(class_list)
y
```

##### Features/Inputs that would be feed to the model

```python
df.iloc[::,:-1]
```

### Feature Scaling

```python
### Using StandardScaler technique to scale all the features
```

```python
from sklearn.preprocessing import StandardScaler
fit=StandardScaler()
X=fit.fit_transform(np.array(df.iloc[:,:-1],dtype=float))
```

<a id='build'></a>

### Split Data into Training and Testing and Building the model

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)
```

```python
from tensorflow.keras.models import Sequential
import pickle
```

```python
def trainModel(model,epochs,optimizer):
    batch_size=128
    model.compile(optimizer=optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    return model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=epochs,batch_size=batch_size)
#     pickle.dump(model, open('./model_pic', 'wb'))
    
```

```python
def plotValidate(history):
    print("Validation Accuracy",max(history.history["val_accuracy"]))
    pd.DataFrame(history.history).plot(figsize=(12,6))
    plt.show()
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(512,activation='relu',input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Dense(256,activation='relu'),
    keras.layers.Dropout(0.2),
    
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Dense(10,activation='softmax'),
])

print(model.summary())
model_history=trainModel(model=model,epochs=600,optimizer='adam')
```

<a id='performance'></a>

## Model Performance

```python
model.save("C_k_d.h5")
```

```python
model_history,model
```

```python
plotValidate(model_history)
```

```python
test_loss,test_acc=model.evaluate(X_test,y_test,batch_size=128)
print("The test loss is ",test_loss)
print("The best accuracy is: ",test_acc*100)
```
